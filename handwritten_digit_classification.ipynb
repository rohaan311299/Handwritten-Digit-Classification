```python
# Importing Libraries
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt 
%matplotlib inline 
import numpy as np
```

```python
# Importing the dataset
(X_train, y_train), (X_test, y_test)=keras.datasets.mnist.load_data()
```

```python
# Scaling the values
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
# flattening the training dataset
X_train_flattened=X_train.reshape(len(X_train),28*28)
X_train_flattened.shape
```

```python
# flattening the test dataset
X_test_flattened=X_test.reshape(len(X_test),28*28)
X_test_flattened.shape
```

```python
X_train_flattened[0]
```

```python
# Neural Network Model
# dense means all inputs are connected to all outputs
model=keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,) ,activation="sigmoid")
])

model.compile(
    optimizer="adam", 
    loss="sparse_categorical_crossentropy", 
    metrics=['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
# evaluating the model
model.evaluate(X_test_flattened, y_test)
```

```python
plt.matshow(X_test[1])
```

```python
# making predictions
y_predicted=model.predict(X_test_flattened)
y_predicted[1]
```

```python
# using argmax to get output with maximum accuracy
np.argmax(y_predicted[1])
```

```python
# doing the same as above for all the predicted values
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
# creating a confusion matrix
cm=tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
# visualizing confusion matrix
# check the error here
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
# with 1 hidden layer

model=keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,) ,activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])

model.compile(
    optimizer="adam", 
    loss="sparse_categorical_crossentropy", 
    metrics=['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
# with 2 hidden layer

model=keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,) ,activation="relu"),
    keras.layers.Dense(100, input_shape=(784,) ,activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])

model.compile(
    optimizer="adam", 
    loss="sparse_categorical_crossentropy", 
    metrics=['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
cm=tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
# Using keras's Flatten function 
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])

model.compile(
    optimizer="adam", 
    loss="sparse_categorical_crossentropy", 
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=5)
```
